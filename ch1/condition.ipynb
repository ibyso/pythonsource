````python
# 주차장 프로그램
# 사용자에게 메뉴 저공
# [1] 주차, [2] 출차, [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상택 ==> ['A']
# 1
# B 자동차 입차. 주차장 상태 ==> ['A','B']
# 2
# B 자동차 출차. 주차장 상태 ==> ['A']

car = ['A','B','C','D','E']
parking_lot = []
top,car_name = 0, 'A'
while True :
    no = input("[1] 주차 | [2] 출차 | [3] 종료")
    if no == "1" :
        if len(parking_lot) == 5 :
            print("주차장 만차")
            continue
        
        parking_lot.append(car[len(parking_lot)])
        print(f"자동차 입차. 주차장 상태 => {parking_lot}")
    elif no == "2" :
        if len(parking_lot) == 0 :
            print("주차장 공차")
            continue

        parking_lot.pop()
        print(f"자동차 출차. 주차장 상태 => {parking_lot}")
    elif no == "3" :
        print("프로그램 종료")
        break
    else :
        print("번호를 확인해 주세요")
        continue
````

````python
parking_lot = []
top,car_name = 0, 'A'
while True :
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    if 0 < no <= 3 :
        if no == 1 :
            if top >= 5 :
                print("주차장 만차")
                continue
            
            parking_lot.append(car_name)
            car_name = chr(ord(car_name)+1)
            top += 1
            print(f"자동차 입차. 주차장 상태 => {parking_lot}")
        elif no == 2 :
            if top <= 0 :
                print("주차장 공차")
                continue

            parking_lot.pop()
            car_name = chr(ord(car_name)-1)
            top -= 1
            print(f"자동차 출차. 주차장 상태 => {parking_lot}")
        elif no == 3 :
            print("프로그램 종료")
            break
    else :
        print("번호를 확인해 주세요")
        continue
````

````python
import random

# 야구게임
target = random.sample(range(1, 10), 3)
chance = 10
# print(numbers)
while True :
    guess = input("숫자 3개 입력(ex:123)")
    selects = [int(i) for i in guess]
    print(selects)
    home = 0
    ball = 0
    out = 0
    for idx, select in enumerate(selects) :
        if target[idx] == select :
            home += 1
        elif select in target :
            ball += 1
        else : 
            out += 1

    print(f"{home} 홈런, {ball} 볼, {out} 아웃입니다. ")
    chance -= 1
    if home == 3 :
        print("게임 승리 !! 결과 : ",target)
        break
    if chance == 0 :
        print("게임 패배 : ",target)
        break
````

## 리스트 컴프리헨션
- 리스트 안에 for 문을 포함

````python
# 1~6 숫자를 가지는 리스트 생성

numbers1 = []
for i in range(1,7) :
    numbers1.append(i)
print(numbers1)

numbers2 = list(range(1,7))
print(numbers2)
````

````python
# 새로운 리스트 생성 : 리스트 컴프리헨션

numbers = [i for i in range(1,7)]
print(numbers)

print([i for i in range(1, 7)])
print([i * i for i in range(5)])
print([i * 2 for i in [1,3,7,9,16]])
print([[i, i+1, i+2] for i in [1,2,3]])

# 1~10 사이의 숫자 중에서 홀수 리스트 생성
odd_list = []
for i in range(1,11) :
    if i % 2 == 1 :
        odd_list.append(i)
print(odd_list)

odd_list2 = [i for i in range(1,11) if i % 2 == 1]
print(odd_list2)
````

````python
q_list = ["갑","을","병","정","경"]

# q_list에서 정을 제외한 요소를 갖는 새로운 리스트 생성

q_list1 = []
for i in q_list :
    if i != "정" :
        q_list1.append(i)
print(q_list1)

q_list2 = [i for i in q_list if i != "정"]
print(q_list2)
````
