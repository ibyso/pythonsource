`숫자가 짝수면 '짝수'라고 출력, 아니면 '홀수'라고 출력`

```python
import random
num = random.randint(1, 100)

print(num)

# if문 : bool 자료형
if num % 2 :
    print("홀수")
else :
    print("짝수")
```

```python
# >, <, >=, <=, ==, != : True, False
# in, not in
a = 200
# a가 100보다 작다면 'a는 100보다 작다' 출력
if a < 100 :
    print("a는 100보다 작다")
else :
    print("a는 100보다 크다")

# a가 100보다 크고 200 이하라면 'a는 100과 200 사이의 수' 출력
if 100 < a <= 200 :
    print("a는 100과 200 사이의 수")
else :
    print("a는 100과 200 사이의 수가 아님")
```

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70 이상이면 C, 나머지는 F

score = 99
if score >= 90 :
    print("A")
elif score >= 80 :
    print("B")
elif score >= 70 :
    print("C")
else :
    print("F")
```

```python
a,b,c = 12, 6, 19

# 가장 큰 숫자 출력
if a >= b and a >= c :
    print("A")
elif b >= a and b >= c :
    print("B")
else :
    print("C")

max_val = a
if max_val < b :
    max_val = b
elif max_val < c :
    max_val = c
print("가장 큰 수",max_val)

# 가장 작은 숫자 출력
min_val = a
if min_val > b :
    min_val = b
elif min_val > c :
    min_val = c
print("가장 작은 수",min_val)
```

```python
fruits = ['딸기','사과','바나나','배','수박']

if '딸기' in fruits :
    print("딸기가 포함됨")
else : 
    print("딸기가 없음")

fruits.clear()
if fruits : 
    print("리스트")
else :
    print("빈 리스트")

if len(fruits) == 0 :
    print("빈 리스트")

if not fruits :
    print("빈 리스트")
```

```python
# and : x and y (x,y가 모두 참인 경우)
# or : x or y (x 또는 y가 참인 경우)
# not x : 반대로

# socre가 90 이상이고, grade가 A인 경우 합격, 나머지 불합격
score, grade = 85, "A"

if score > 90 and grade == "A" :
    print("합격")
else : 
    print("불합격")

# socre가 90 이상이거나, grade가 A인 경우 합격, 나머지 불합격
if score > 90 or grade == "A" :
    print("합격")
else : 
    print("불합격")
```
