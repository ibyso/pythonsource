```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str__(self):
        return "허용되지 않은 별명입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)

# say_nick("야호")
# say_nick("바보")


try:
    say_nick("바보")
except MyError as e:
    print(e)
```

```python
# 문제 1. 랜덤 비밀번호 생성기
# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# choice()
chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789!@#$%"

# 비밀번호 길이는 4 이상만 허용
# 숫자가 아닌 값을 입력하면 예외처리

import random


while True :
    try:
        pw_len = int(input("비밀번호 길이 입력"))
        if pw_len < 4 :
            print("비밀번호는 4자리 이상이어야 합니다.")
            continue
        break
    except ValueError:
        print("숫자만 입력하세요")

pw = ""
for s in range(pw_len) :
    pw += random.choice(chars)
print(f"비밀번호는 {pw}")

# password = []
# for s in range(pw_len) :
#     password.append(random.choice(chars))
password = [random.choice(chars) for _ in range(pw_len)]
print(f"비밀번호는 {"".join(password)}")
```

```python
# 문제 2.나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2001-08-25(datetime.strptime("2001-01-01","%Y-%m-%d"))
from datetime import datetime

try:
    birth = datetime.strptime(input("날짜 입력(ex.2001-01-01)"),"%Y-%m-%d")
    now = datetime.now()
    age = now.year - birth.year
    if (now.month, now.day) < (birth.month, birth.day) :
        age -= 1

    print(f"나이는 {age}세")

except ValueError:
    print("올바른 형식의 날짜를 입력하세요")
```

```python
# 문제 3. 학생 정보를 json으로 저장하기
# 학생 정보를 입력받아 json 형식으로 파일 저장
# 이름, 나이, 전공
# 읽어서 출력
import json
try:
    subject =["이름","나이","전공"]
    profile = {}
    for e in subject :
        profile[e] = input(e)

    with open("./data/profile1.json","w",encoding="utf-8") as f :
        json.dump(profile, f,ensure_ascii=False)

    with open("./data/profile1.json","r",encoding="utf-8") as g :
        print(json.load(g))
except FileNotFoundError as e:
    print("파일 오류", e)
except :
     print("오류")
```

```python
# 다음과 같은 csv 파일이 있다
# 이름, 국어, 영어 수학
# 홍길동 90, 80, 70
# ...

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0

#csv.DictReader() 사용, 평균은 둘째자리까지

import csv
#########
## csv 파일 생성
#########
with open("./data/sampleTest1.csv","w",encoding="utf-8",newline="") as f :
    headers = ['이름','국어','영어','수학']
    # header 작성
    writer = csv.DictWriter(f, fieldnames=headers)
    writer.writeheader() # 헤더 삽입

    writer.writerow({"이름":"홍길동","국어":88,"영어":92,"수학":94})
    writer.writerow({"이름":"성춘향","국어":87,"영어":90,"수학":99})
    writer.writerow({"이름":"아무개","국어":91,"영어":94,"수학":86})
```

```python
#########
## csv 문제 작성
#########
import csv
try:
    with open("./data/sampleTest1.csv","r",encoding="utf-8") as f :
        reader = csv.DictReader(f)
        for c in reader :
                avg = (int(c["국어"]) + int(c["영어"]) + int(c["수학"])) / 3
                print(f"{c.get("이름")}의 평균 점수 : {avg:.2f}")
except FileNotFoundError as e:
    print(e)
except :
    print("에러")
```
