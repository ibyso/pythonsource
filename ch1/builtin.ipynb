```python
# abs() : 절대값 리턴

print(abs(3))
print(abs(-3))
print(abs(3.12))
```

```python
# chr() <=> ord()
# chr() : 유니코드 값을 넣으면 그 코드에 해당하는 문자를 리턴
# ord() : 문자를 넣으면 문자에 해당하는 유니코드를 리턴

print(chr(97))
print(chr(44032))
print(ord('a'))
print(ord('가'))
```

```python
# dir() : 객체가 지닌 변수나 함수를 보여주는 함수

list1 = [1,2,3]
dir(list1)

import math
dir(math)
```

```python
# divmod(a, b) : 몫, 나머지를 튜플로 반환

quot, rem = divmod(7, 3)
print(quot,rem)
```

```python
# enumerate() : 리스트, 튜플, 문자열을 입력으로 받아 인덱스 값을 포함하는 enumerate 객체 반환

for i, name in enumerate(['body','foo','bar'],1):
    print(i,name)

for i, name in enumerate('bodybar',start=1):
    print(i,name)
```

```python
# zip() : 동일한 개수로 이루어진 데이터들을 묶어서 반환해주는 함수

print(list(zip([1,2,3],[4,5,6])))
print(list(zip([1,2,3],[4,5,6],[7,8,9])))
print(list(zip("abc","def")))
```

```python
# 학생 5명의 점수를 입력받아 다음 정보를 출력하는 프로그램을 작성

# 출력내용
# 최고점수(max), 최저점수(min), 평균점수(소수 2째자리)(sum/len), 점수를 오름차순 출력(sorted), 등수와 출력(enumerate)

score = []
cnt = 5
while cnt > 0 :
    try :
        score.append(int(input("점수 입력 : ")))
        cnt -= 1
    except :
        print("숫자를 입력하세요")

print(f"최고 점수 : {max(score)}")
print(f"최저 점수 : {min(score)}")
print(f"평균 점수 : {round(sum(score) / len(score),2)}")
print(f"점수(오름차순): {sorted(score)}")
for idx,s in enumerate(sorted(score,reverse=True)) :
    print(f"등수 : {idx+1}, 점수 : {s}")
```

```python
# 문자열을 입력받아 다음 정보를 출력
# 출력 내용 : 입력한 문자, 각 문자의 유니코드 값, 가장 큰 문자, 가장 작은 문자

# 문자열 입력 :Python
# P -> 80
# y -> 121
#
# 가장 큰 문자 : y
# 가장 작은 문자 : P

sentence = input("문자열 입력")

for s in sentence :
    print(f"{s} -> {ord(s)}")

print(f"가장 큰 문자 : {max(sentence)}")
print(f"가장 작은 문자 : {min(sentence)}")
```

```python
x = [3,7,1,-5,9]
y = [2,-4,8,10,-1]

# zip, abs, enumerate
# 출력
# (3,2) -> 5

# list1 = list(zip(x,y))
# for a,b in list1 :
#     print(f"({a},{b}) -> {abs(a+b)}")

for i, (a,b) in enumerate(zip(x,y)) :
    distance = abs(a) + abs(b)
    print(f"{a},{b} -> {distance}")
```

```python
# 사용자로부터 숫자 10개 입력받기

# 출력
# 입력리스트, 중복제거, 오름차순, 튜플변환, 집합 크기

# 원본 [3,5,2,5...]
# 중복제거 {1,2,3,4...} set()
# 정렬 sorted()
# 튜플 tuple()
# 집합 크기 : len()

list2 = []
cnt = 10
while cnt > 0 :
    try :
        list2.append(int(input("숫자 입력 : ")))
        cnt -= 1
    except :
        print("숫자를 입력하세요")

list_set = set(list2)
print(f"원본 : {list2}")
print(f"중복 제거 : {list_set}")
print(f"정렬 : {sorted(list_set)}")
print(f"튜플 : {tuple(sorted(list_set))}")
print(f"집합크기 : {len(list_set)}")
```
