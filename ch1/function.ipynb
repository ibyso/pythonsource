````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 매개변수
# 입력값이 몇 개가 될지 모르는 경우
# *

# def 함수명(*변수명) :
#   해야할 일

def add(*args) : 
    result = 0
    for i in args :
        result = result + i
    print(result)
````

````python
add(2,3)
add(2,3,4,5)
add(2,3,4,5,6,7)
````

````python
# 가변 매개변수와 혼용 할 때
# def add_mul(choice, *args, a) :
#     pass

# add_mul("add",10,11,12,13,a=10)


# 일반 매개변수는 앞쪽으로 선언
def add_mul(choice, *args) :
    if choice == "add" :
        result = 0
        for i in args :
            result += i
    elif choice == "mul" :
        result = 1
        for i in args :
            result *= i
    else :
        result = -1

    return result
````

````python
print(add_mul("add",10,11,12,13))
print(add_mul("mul",10,11,12,13))
````

````python
# 키워드 매개변수 (key=value)
# **변수명

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs) :
    print(kwargs)

    # key 값만 출력
    print(kwargs.keys())

    # value 값만 출력
    print(kwargs.values())

    # key,value 같이 출력
    for k,v in kwargs.items() :
        print(k,v)
````

````python
print_kwargs(name="kim",age=15)
print_kwargs(name="park",age=25, addr="서울")
print_kwargs(name="hong",age=15, addr="경기", tel="010-1234-5678")
````

````python
# 변수의 효력 범위

# 함수 밖의 변수 : 전역변수
a = 1
def vartest(a) :
    # 매개변수 a : 지역변수
    a = a +1

vartest(a)
print(a)
````

````python
# 전역변수 a의 값을 변경하기
# 리턴 이용(추천 방법)

a = 1
def vartest(a) :
    # 매개변수 a : 지역변수
    a = a +1
    return a

a = vartest(a)
print(a)
````

````python
# global 키워드 사용
# 함수 안에서 전역변수 접근 안 됨

a = 1
def vartest() :
    global a
    a = a +1

vartest()
print(a)
````

````python
# 입력으로 숫자(자연수)를 받아서 자연수가 짝수라면 True, 아니면 False 리턴
# is_even()
def is_even(a) : 
    return a % 2 == 0

print(is_even(1))
print(is_even(4))
print(is_even(66))
print(is_even(71))

def print_even(a) :
    if a == True :
        return "짝수"
    else :
        return  "홀수"

print(print_even(is_even(1)))
print(print_even(is_even(4)))
print(print_even(is_even(66)))
print(print_even(is_even(71)))
````

````python
# 평균 구하는 함수
# get_average()
# 요소의 개수는 정해지지 않음
# 리턴 없음

def get_average(*args) :
    print(sum(args)/len(args))

get_average(10,11,12)
get_average(10,11,12,15,25)
get_average(10,11,12,15,25,36)

````

````python
# 사칙연산 함수
# 결과 리턴

def calculator(sign, a, b) : 
    if sign == "+" :
        return a + b
    elif sign == "-" :
        return a - b
    elif sign == "*" :
        return a * b
    elif sign == "/" :
        return a / b
    else :
        return
    
print(calculator("+",15,3))
print(calculator("-",15,3))
print(calculator("*",15,3))
print(calculator("/",15,3))
print(calculator("a",15,3))
````

````python
# filter()
list1 = list(range(1,21))

# filter(함수, iterable)

# 짝수만 새로운 리스트로 생성
# 방법 1
even_list1 = []
for i in list1 :
    if i % 2 == 0 :
        even_list1.append(i)

# 방법 2 (리스트 컴프리헨션)
even_list2 = [i for i in list1 if i % 2 == 0]

# 방법 3 (filter)
even_list3 = list(filter(lambda x : x%2 == 0,list1))
print(even_list1)
print(even_list2)
print(even_list3)
````

````python
strings = ['Bob','charles','Alexander3','teddy','apple','Banana']

# 단어의 시작이 대문자인 단어들만 모아서 새로운 리스트로 생성

list1 = []
for i in strings :
    if i[0].isupper() :
        list1.append(i)
print(list1)

list2 = list(i for i in strings if i[0].isupper())
print(list2)

# filter(함수,원본 리스트) : 새로운 리스트로 생성
list3 = list(filter(lambda x : x[0].isupper(),strings))
print(list3)
````

````python
list1 = [i for i in range(1,21)]

# 3의 배수만 모아서 새로운 리스트로 생성

# 리스트 컴프리헨션
list2 = [i for i in list1 if i % 3 == 0]
print(list2)

# list(filter())
list3 = list(filter(lambda x : x % 3 == 0, list1))
print(list3)
````

````python
# map(function, iterable)
# 전체 요소 대상으로 함수 적용
map1 = list(map(lambda x : x % 3 == 0, list1))
map1
````
