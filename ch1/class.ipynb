```python
class Calculator :
    def __init__(self) :
        self.result = 0

    def add(self,num) :
        self.result += num
        return self.result

    def sub(self,num) :
        self.result -= num
        return self.result
```

```python
# 객체 생성 => __init()__ 자동으로 호출
# 변수 = 클래스명()

calc1 = Calculator()
print(calc1.add(3))
print(calc1.add(4))

calc2 = Calculator()
print(calc2.add(13))
print(calc2.add(14))
```

#### 인스턴스(멤버) 변수
- 객체마다 따로따로 가지는 데이터
- `self.변수명` 형태로 작성

```python
# 자동차 클래스 설계
# 속성 : 색상, 속도 => 멤버 변수
# 동작 : 전진, 후진, 가속, 감속 => 멤버 메소드

class Car :
    def __init__(self, color, speed):
        # 인스턴스(멤버) 변수
        self.color = color
        self.speed = speed

    # 인스턴스(멤버) 메소드(함수)
    def upSpeed(self, value) :
        self.speed += value

    def downSpeed(self, value) :
        self.speed -= value

    def __str__(self):
        return f"Car(color : {self.color},\tspeed : {self.speed})"

    
```

```python
# Car 객체 생성
# 변수명 = 클래스명()

# __init__() 자동으로 호출
car1 = Car("Red",50)
car2 = Car("Black",70)

print(car1)
print(car2)
```

```python
# 인스턴스 매서드 호출
# 인스턴스 변수 값을 변경할 수 있음
car1.upSpeed(30)
print(car1)

car2.downSpeed(10)
print(car2)
```

#### 정적 메서드
- `self`, `cls` 도 필요없는 클래스와 논리적으로만 관련된 매서드
- 객체의 상태나 클래스 변수를 전혀 건드리지 않을 때 사용

```python
class Math :
    @staticmethod
    def add(a,b) :
        return a + b
    
# math1 = Math()
# math1.add(5,6)

# 클래스명.매서드명()
Math.add(10,20)
```

| 구분 | 첫 인자 | 접근 가능 | 주 용도 |
| --- | ------ | -------- | ------ |
|`인스턴스 메서드`|`self`| 인스턴스 변수, 클래스 변수 모두 | 객체별 동작 |
|`@classmethod`|`cls`| 클래스 변수만 | 클래스 단위 동작 |
|`@staticmethod`| 없음 | 없음 | 클래스와 관련은 있지만 독립적인 계산 |

```python
class Student :
    # 클래스 변수
    school_name = "솔데스크"
    student_count = 0

    def __init__(self,name):
        self.name = name
        # 전체 학생 수
        Student.student_count += 1

    # 인스턴스 매서드
    def show_info(self):
        print("=== 인스턴스 매서드 ===")
        print(f"학생 이름 : {self.name}") # 인스턴스 변수 접근
        print(f"학교명 : {Student.school_name}") # 클래스 변수 접근
        print(f"학생 수 : {Student.student_count}")
    
    @classmethod
    def show_school_info(cls):
        print("=== 클래스 매서드 ===")
        # print(f"학생 이름 : {self.name}") # 인스턴스 변수 접근
        print(f"학교명 : {Student.school_name}") # 클래스 변수 접근
        print(f"학생 수 : {Student.student_count}")

    @staticmethod
    def add_score(score1,score2):
        print("=== 정적 매서드 ===")
        # print(f"학생 이름 : {self.name}") # 인스턴스 변수 접근
        print(f"학교명 : {Student.school_name}") # 클래스 변수 접근
        print(f"학생 수 : {Student.student_count}")
        print(f"총점 :{score1+score2}")


stu1 = Student("홍길동")
stu2 = Student("성춘향")

# 인스턴스 매서드 호출
stu1.show_info()
stu2.show_info()

# 클래스 매서드 호출
Student.show_school_info()

# 정적 매서드 호출
Student.add_score(80,95)
```

```python
# Student 클래스 작성
# 멤버변수 : 이름(name), 나이(age), 점수(score)
# 메서드 : student(hour) : 공부한 시간 * 2 = 결과를 점수에 추가
#         showInfo() : 이름 : 홍길동, 나이 : 25, 점수 : 75

class Student :
    def __init__(self, name, age, score):
        self.name = name
        self.age = age
        self.score = score

    def student(self,hour) :
        self.score += hour*2
    
    def showInfo(self):
        print(f"이름 : {self.name}, 나이 : {self.age}, 점수 :{self.score}")


stu1 = Student("홍길동",25,75)
stu2 = Student("성춘향",24,85)

stu1.showInfo()
stu2.showInfo()

stu1.student(10)
stu2.student(5)

stu1.showInfo()
stu2.showInfo()
```

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
# 매서드 : deposit(money) : 잔액 + 입금액
#         withdraw(money) : 잔액 - 출금액
#         showBalance() : 현재 잔액 출력

class Account :
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self,money) :
        self.balance += money

    def withdraw(self,money) :
        if self.balance < money :
            print("잔액 부족")
        else : 
            self.balance -= money

    def showBalance(self) :
        print(f"계좌주 : {self.owner}, 잔액 : {self.balance}원")


acc1 = Account("가나다",25000)
acc2 = Account("abc",55000)

acc1.showBalance()
acc2.showBalance()

acc1.deposit(10000)
acc2.withdraw(800000)
acc2.withdraw(8000)

acc1.showBalance()
acc2.showBalance()
```

```python
# Order 클래스 작성
# 클래스 변수 : total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
# 매서드 : getTotalPrice() = 가격 x 수량 : 총 주문금액 반환
# showOrder : 출력문 작성
# 메뉴 : 아메리카노 가격 : 4000원 수량 : 2개 총 주문금액 : 8000원

class Order :
    total_order = 0

    def __init__(self,menu,price,count):
        self.menu = menu
        self.price = price
        self.count = count
        Order.total_order += 1
    
    def getTotalPrice(self) :
        # Order.total_order += self.price * self.count
        return self.price * self.count
    
    def showOrder(self) :      
        print(f"메뉴 : {self.menu}, 가격 : {self.price}, 수량 : {self.count}, 총 주문금액 : {self.getTotalPrice()}")


order1 = Order("아메리카노",3000,2)
order2 = Order("케이크",8000,1)

order1.showOrder()
order2.showOrder()
print(f"전체 주문 수 : {Order.total_order}")
```

#### 상속
- 기존 클래스(부모/상위)의 멤버변수와 매서드를 새로운 클래스(지식/하위)가 그대로 물려받는 것
- 코드 재사용성, 확장

```python
class Parent:
    def __init__(self):
        self.value="테스트"
        print("Parent 클래스의 __init__ 호출")
    
    def test(self) :
        print("Parent 클래스의 test() 호출")

class Child(Parent):
    pass

child = Child()
child.test()
print(child.value)
```

```python
# super() : 상위 클래스

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child 클래스의 __init__ 호출")

child = Child()
child.test()
print(child.value)
```

```python
class Animal :
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name}가 먹이를 먹는다")

    def sound(self):
        print(f"{self.name}가 소리를 낸다")

class Dog(Animal) :
    def __init__(self, name, kind):
        super().__init__(name)
        self.kind = kind

    def fetch(self) :
        print(f"{self.name}이 공을 물어옵니다")


dog1 = Dog("초코","푸들")
dog1.eat()
dog1.sound()
dog1.fetch()
```

```python
# Audio : 멤버변수 : power, volume  / 메소드 : 전원 조절, 볼륨 조절 + tune()
# TV : 멤버변수 : power, volume, size   / 메소드 : 전원 조절, 볼륨 조절 + watch()

class AudioVisual :
    def __init__(self, power, volume):
        self.power = power
        self.volume = volume
    
    def switch(self, on_off) :
        self.power = on_off

    def setVolume(self,volume):
        self.volume = volume
    
class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        print("LaLaLa" if self.power else "turn it on")

class TV(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size
    
    def watch(self):
        print("have fun!" if self.power else "switch on")


tv = TV(False, 12 ,50)
# tv 켜기
tv.watch()
tv.switch(True)
tv.watch()

audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
```

#### 오버라이딩
- 부모 클래스에 있는 메서드를 자식 클래스에서 같은 이름으로 다시 정의하여 동작 내용 변경

```python
class Animal:
    def __init__(self,name):
        self.name = name
    
    def sound(self):
        print(f"{self.name}가 소리를 냅니다")

class Dog(Animal):
    def sound(self):
        # return super().sound()
        print(f"{self.name} : 멍멍")
    
class Cat(Animal):
    def sound(self):
        # 오버라이딩 시 동일한 부모 매서드를 호출하고 싶다면
        super().sound()
        print(f"{self.name} : 야옹")

dog1 = Dog("초코")
cat1 = Cat("몰리")

dog1.sound()
cat1.sound()
```

```python
# Employee 부모 클래스 :
# 멤버 변수 : 이름(name), 급여(pay)
# 메소드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
# 메소드 : getTotalPay() : 급여 + 보너스 출력


class Employee:
    def __init__(self,name,pay):
        self.name = name
        self.pay = pay
    
    def showInfo(self) :
        print(f"이름 : {self.name}, 급여 : {self.pay}")
    
class Manager(Employee):
    def __init__(self, name, pay, bonus):
        super().__init__(name, pay)
        self.bonus = bonus
    
    def getTotalPay(self) :
        print(f"이름 : {self.name}\n총 급여 : {self.pay + self.bonus}")

emp = Employee("홍길동",180)
emp.showInfo()

manager1 = Manager("아무개",200,50)

manager1.showInfo()
manager1.getTotalPay()
```

```python
# Shape 부모 클래스
# 매서드 : area() => "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 매서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : 반지름(radius)
# 매서드 : area() => 원의 넓이 출력

class Shape :
    def area() :
        print("도형의 넓이")

class Rectangle(Shape):
    def __init__(self,width,height):
        super().__init__()
        self.width = width
        self.height = height
    
    def area(self):
        print(f"직사각형의 넓이 : {self.width * self.height}")

class Circle(Shape):
    def __init__(self,radius):
        super().__init__()
        self.radius = radius
    
    def area(self):
        print(f"원의 넓이 : {self.radius **2 * 3.14}")

rect = Rectangle(10,20)
c = Circle(5)

rect.area()
c.area()
```
