## 5. 딕셔너리

- {}
- {key:value}

```python
dict1 = {}
dict2 = {"name":"hong","phone":"010-9999-1234","birth":"1181"}
dict3 = {0:"python",1:[1,2,3,4,5],1:(6,7,8)}

def reset_dict() :
    dict1 = {}
    dict2 = {"name":"hong","phone":"010-9999-1234","birth":"1181"}
    dict3 = {0:"python",1:[1,2,3,4,5]}
    return dict1, dict2, dict3

```

```python
# 조회
dict1["role"] = "test"
dict2["name"] = "kong"

print(dict1)
print(dict2["name"])
print(dict3[1])

dict1, dict2, dict3 = reset_dict()
print(dict1)
print(dict2["name"])
print(dict3[1])
```

```python
# 추가
dict2['addr'] = "서울시 종로구"
dict2
```

```python
# 2 : "hello,python","hello java"
dict3[2] = ["hello,python","hello java"]
print(dict3)
```

```python
del dict3[1]
print(dict3)
```

```python
# 딕셔너리 내 모든 키 값 가져오기
print(dict2.keys())

# 딕셔너리 내 모든 값 가져오기
print(dict2.values())

# 딕셔너리 내 모든 쌍(key,value) 가져오기
print(dict2.items())
```

```python
list1.clear()
print(list1)

dict2.clear()
print(dict2)
```

```python
# dict3[1] Error
print(dict3.get(1)) # none
print(dict3.get(2))
```

```python
# in : 리스트에서 해당 요소가 존재하는지 여부
print('z' in list2)
print(2 in t4)
print(0 in dict3)   # key를 검색
print('e' in "hello")

print('z' not in list2)
print(2 not in t4)
print(0 not in dict3)   # key를 검색
print('e' not in "hello")
```

```python
# 실습

dict1 = {'A':90,'B':80,'C':70}
print(dict1['B'])

dict2 = {"성인":20000,"청소년":10000,"아동":5000}
dict2["소아"] = 0
print(dict2)

print(dict2.keys())
print(dict2.values())

# 리스트로 바로 받아보는 방법
print(list(dict2.keys()))
print(list(dict2.values()))
```

## 6. 집합 자료형
- 집합에 관련된 것을 처리하기 위한 자료형
- 중복 제거
- 순서가 없음
- set() : 생성

```python
s1 = set()
s2 = set([1,2,3,4,5,4])
s3 = set([1,2,3,4,5,6])

print(s1)
print(s2)
print(s3)
```

```python
# TypeError: 'set' object is not subscriptable
# s2[0]

# 인덱싱, 슬라이싱을 하고 싶다면 가능한 타입으로 변환
# 리스트 : list(), 튜플 : tuple()

s1 = "안녕하세요"
print(type(s1))
print(type(3))
print(type(s2))
print(type(list3))
print(type(t4))
print(type(dict2))

print(type(list(s2)))

# set 자료형 => 새로운 list로 생성
print(s2)
print(list(s2))
print(s3)
print(tuple(s3))

# 인덱싱
print(list(s2)[2])
print(tuple(s3)[3])
```

```python
# set() : list, tuple set 구조로 변환 시 사용
print(set(list4))
print(set(s1))
```

```python

s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

# 교집합
print(s1 & s2)
print(s1.intersection(s2))

# 합집합
print(s1 | s2)
print(s1.union(s2))

# 차집합
print(s1 - s2)
print(s1.difference(s2))
```

```python
# 값 1개 추가하기
s1.add(8)
print(s1)

# 값 여러개 추가하기
# TypeError: cannot use 'list' as a set element (unhashable type: 'list')
# s1.add([9,10])
s1.update([9,10])
print(s1)
```
