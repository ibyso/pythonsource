```python
# 자료형
# 숫자형, 문자형

# 1. 숫자형 - 정수, 실수, 8진수, 16진수
a = 123
print(f"양의 정수 : {a}")

a = -178
print(f"음의 정수 : {a}")
```

```python
# 연산자
# +, -, *, /, //, %, **

a = 3
b = 4
print(f"a+b = {a+b}")
print(f"a-b = {a-b}")
print(f"a*b = {a*b}")
print(f"a/b = {a/b}")
print(f"a//b = {a//b}") # 몫만
print(f"a%b = {a%b}")   # 나머지만
print(f"a**b = {a**b}")
```

#### 인덱싱과 슬라이싱
> - 인덱스는 0번부터 시작

```python
a = "life is too short, You need Python"
print(a[33])
print(a[-1])
```

```python
# 슬라이싱 시 끝번호는 포함하지 않음

print(a[0:5])
print(a[0:2])
print(a[5:7])
print(a[12:17])
print(a[:])
print(a[19:-7])
print(a[19:])
print(a[:5])
```

```python
a = "20260714Rainy"
# 날짜와 날씨 분리
date = a[:8]
weather = a[8:]

print(f"date : {date}")
print(f"weather : {weather}")

print(f"year : {date[:4]}")
print(f"month : {date[4:6]}")
print(f"date : {date[6:]}")
```

#### 문자열 포매팅
> - 문자열 안에 어떤 값을 삽입하는 방법  
  
| 코드 | 설명 |
| --- | --- |
| `%s` | 문자열(string) |
| `%d` | 정수(inteager) |
| `%c` | 문자 1개(character) |
| `%f` | 부동 소수(floating-point) |
| `%o` | 8진수 |
| `%x` | 16진수 |
| `%%` | %문자 |  
  

> - 
> - format() 함수 사용
> - f 문자열 포매팅(파이썬 3.6버전부터 지원)

```python
a = "I eat %d apples" % 3
a
```

```python
# %s => 어떤 값이든 대입 가능

a = "I eat %s apples" % "five"
a
```

```python
a = "Error is %d%%" % 98
a
```

```python
a = "I ate %d apples. so I was sick for %s days" % (10, "three")
a
```

```python
# 오른쪽 정렬
print("%10s" % "hi")
# 왼쪽 정렬
print("%-10sJane" % "hi")

# 0 : 길이에 상관하지 않음
# 소수점
print("%0.4f" % 3.141592)

# 전체 길이 10인 문자열 공간에서 오른쪽 정렬
print("%10.4f" % 3.141592)
```

```python
a = "I eat {} apples".format(3)
print(a)

a = "I ate {} apples. so I was sick for {} days".format(10, "three")
print(a)

a = "I ate {0} apples. so I was sick for {1} days".format(10, "three")
print(a)

a = "I ate {1} apples. so I was sick for {0} days".format(10, "three")
print(a)
```

```python
name = "홍길동"
age = 30

# 
print("그의 이름은 %s이고, 나이는 %d 입니다" % (name,age))
print("그의 이름은 {}이고, 나이는 {} 입니다".format(name,age))
print(f"그의 이름은 {name}이고, 나이는 {age+1} 입니다")
```

```python
# 대소문자 변경

print("Hello".upper())
print("Hello".lower())
```

```python
# 공백 제거

a = "hello"
b = "h e l l o"
c = "    hello    "
print(f"왼쪽 공백 제거 : [{c.lstrip()}]")
print(f"오른쪽 공백 제거 : [{c.rstrip()}]")
print(f"양쪽 공백 제거 : [{c.strip()}]")
print(f"양쪽 공백 제거 : [{b.strip()}]")
print(f"비교 : [{ a == c.strip()}]")

print(f"체크 : [{c}]")
```

```python
# 문자열 변경 : replace("바뀔 문자열","변경할 문자열")

a = "Life is too short"
print(f"문자열 변경 : [{a.replace("short","long")}]")
print(f"체크 : [{a}]")
```
