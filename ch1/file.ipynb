# file 읽고 쓰기

| 파일 열기 모드 | 설명 |
| ------------ | ---- |
| `r` | **읽기(Read) 모드** : 파일의 내용을 읽기만 할 때 사용. 파일이 존재하지 않는 경우 오류 발생. |
| `w` | **쓰기(Write) 모드** : 파일의 내용을 새로 쓸 때 사용. 기존 파일이 존재하는 경우 내용 삭제 후 새로 작성. 없는 경우 새로 생성. |
| `a` | **추가(Append) 모드** : 기존 파일의 마지막에 내용을 추가할 때 사용. 파일이 없는 경우 새로 생성. |

```python
# score.txt 내용을 읽은 후 평균을 구한 후 출력
scores = []
with open("./data/score.txt","r") as f :
    for num in f :
        scores.append(int(num))
print(scores)

avg = sum(scores) / len(scores)
print(f"평균 : {avg}")
```

```python
# sample.txt 파일을 읽어서 총합과 평균을 구한 후 총합과 평균을 result.txt 파일 작성

samples = []
with open("./data/sample.txt","r") as f :
    for num in f :
        samples.append(int(num))
print(samples)

with open("./data/result.txt","w",encoding="UTF-8") as f :
    f.write(f"총합 : {sum(samples)}\n")
    f.write(f"평균 : {(sum(samples)/ len(samples)):.2f}\n")
```

```python
# 1000명의 이름, 키, 몸무게 info.txt를 읽은 후
# BMI 지수 계산

# bmi = 체중 / 키의 제곱
# bmi가 18.5 미만일 때 저체중
# bmi가 18.5 ~ 22.9 정상체중
# bmi가 23.0 ~ 24.9 과체중
# bmi가 25 ~ 비만

# 출력문
# 이름 : 하타
# 몸무게 : 95
# 키 : 180
# BMI : 18.6
# 결과 : 정상체중

infos = []

with open("./data/info.txt","r",encoding="utf-8") as f :
    for line in f :
        name, weight, height = line.strip().split(", ")
        weight, height = int(weight), int(height)
        bmi = round(weight / ((height/100)**2),1)
        result = ""
        if bmi < 18.5 :
            result = "저체중"
        elif 18.5 <= bmi < 23 :
            result = "정상체중"
        elif 23 <= bmi < 25 :
            result = "과체중"
        elif bmi >= 25 :
            result = "비만"
        
        # print("\n".join(["이름:{}", "체중:{}","키:{}","BMI:{}","결과:{}"]).format(name,weight,height,bmi,result))
        # print()
        infos.append({"이름":name,"체중":weight,"키":height,"BMI":bmi,"결과":result})

with open("./data/info_result.txt","w",encoding="utf-8") as g :
    for info in infos :
        for idx, obj in info.items() :
            g.write(f"{idx} : {obj}\n")
        g.write("\n")
```

```python
# origin.txt 읽어서 암호화 시킨 후 encr.txt 저장
# 1. 암호화 : origin.txt 읽어서 변수에 담기 => 그걸 암호화 encr.txt에 저장
# 2. 복호화 : 암호화된 파일을 읽은 후 원래대로 출력

encr_key = 123
while True :
    # 입력을 받는다
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 중 원하는 메뉴의 번호를 입력하세요."))

    # 입력받은 값으로 구분한다
    if no == 1 :
        # 파일을 열어서 입력을 받는다
        with open("./data/origin.txt","r",encoding="utf-8") as f :
            origin = f.read()
            encr = ""
            for i in origin :
                encr = encr+chr(ord(i)+encr_key)

        with open("./data/encr.txt","w",encoding="utf-8") as g :
            g.write(encr)

    elif no == 2 :
        with open("./data/encr.txt","r",encoding="utf-8") as f :
            encr = f.read()
            origin = ""
            for i in encr :
                origin = origin+chr(ord(i)-encr_key)

            print(origin)

        # with open("./data/origin.txt","w",encoding="utf-8") as g :
        #     g.write(encr)
    elif no == 3 :
        break
    else :
        break
```
