## 사용자 입출력 함수
- input()
- print()

```python
# 사용자가 입력한 값을 어떤 변수에 대입하고자 할 때
# 입력값은 무조건 문자
input()
```

```python
def calculator(sign, a, b) : 
    if sign == "+" :
        return a + b
    elif sign == "-" :
        return a - b
    elif sign == "*" :
        return a * b
    elif sign == "/" :
        return a / b
    elif sign == "//" :
        return a // b
    else :
        return

# 숫자 2개 입력받기, 연산자 기호 (+,*,...) 입력 받기
num1 = int(input())
num2 = int(input())
op = input()

print(calculator(op,num1,num2))


# print(calculator("+",15,3))
# print(calculator("-",15,3))
# print(calculator("*",15,3))
# print(calculator("/",15,3))
# print(calculator("//",15,3))
# print(calculator("a",15,3))
```

```python
values = input("두 숫자와 연산자 기호 입력 (ex 15 10 +)")

# 문자(공백을 기준으로 분리)
# '15 10 +'
# result = values.split()
# print(result)
num1, num2, op = values.split()

print(calculator(op, int(num1),int(num2)))
```

```python
values = input("두 숫자와 연산자 기호 입력 (ex 15,10,+)")
num1, num2, op = values.split(",")

print(calculator(op, int(num1),int(num2)))
```
