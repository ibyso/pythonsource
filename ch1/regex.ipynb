# 정규표현식(정규식)
- 복잡한 문자열을 처리할 때 사용하는 방법
- 파이썬만의 고유한 문법이 아니라 문자열을 처리하는 모든 곳에서 사용하는 일종의 형식 언어

```python
# 주민등록번호 => 뒷자리를 모두 * 문자로 변경

# 파이썬에서 정규식 지원하는 모듈
import re
dir(re)

# re.compile() : 정규식을 작성
pattern = re.compile("ab*")

# 패턴을 특정 문자열이 포함하고 있는가?
print(pattern.match("abc"))
```

| 매서드 | 목적 |
| ----- | ---- |
| match | 문자열의 처음부터 정규식과 매칭되는지 조사 |
| search | 문자열 전체를 검색하여 정규식과 매칭되는지 조사 |
| findall | 정규식과 매치되는 모든 문자열(substring)을 리스트로 리턴 |
| finditer | 정규식과 매칭되는 모든 문자열(substring)을 이터레이터 객체로 리턴 |

```python
p = re.compile("[abc]")

print(p.match("a"))
print(p.match("before"))
print(p.match("dude"))
```

```python
m = p.match("a")

print(f"패턴 시작 : {m.start()}")
print(f"패턴 끝 : {m.end()}")
print(f"패턴과 일치하는 문자열 : {m.group()}")
print(f"일치하는 시작, 끝 반환 : {m.span()}")

if m :
    print(f"match found {m.group()}")
else :
    print("no match")
```

```python
m = p.findall("Happy forever")
print(m)

m = p.finditer("Happy forever")
print(m)

for r in m :
    print(r)
```

```python
# . 문자 : \n을 제외한 모든 문자
p = re.compile("a.b")

print(p.match("a\nb"))
print(p.match("abb"))
print(p.match("a0b"))
print(p.match("aab"))
print(p.match("a b"))
```

```python
# . : 마침표로 사용하고 싶을 떄
# \. OR [.]
# p = re.compile("a.b")    "." : escape 문자로 제대로 인식 못함
# p = re.compile("a\.b")
p = re.compile(r"a[.]b")

print(p.match("a\nb"))
print(p.match("abb"))
print(p.match("a0b"))
print(p.match("aab"))
print(p.match("a b"))
print(p.match("a.b"))
```

```python
# * : 0 ~ 무한대
# + : 1 ~ 무한대
# ? : 0 ~ 1

p = re.compile("ca*t")

print(p.match("ct"))
print(p.match("cat"))
print(p.match("caaaaat"))
```

```python
p = re.compile("ca+t")

print(p.match("ct"))
print(p.match("cat"))
print(p.match("caaaaat"))
```

```python
p = re.compile("ca?t")

print(p.match("ct"))
print(p.match("cat"))
print(p.match("caaaaat"))
```

```python
p = re.compile("ca{2}t")

print(p.match("ct"))
print(p.match("cat"))
print(p.match("caat"))
print(p.match("caaaaat"))
```

```python
# a~z 모든 소문자
p = re.compile("[a-z]+")

print(p.search("kkk1234"))
print(p.search("k1234a12b"))
```

```python
# [^ ] : 대괄호 안에 있는 문자를 제외한

p = re.compile("[^a-zA-Z0-9]+")

print(p.search("kkk1234"))
print(p.search("k1234a12b"))
```

```python
p = re.compile("[가-힇]+")

print(p.search("1234 안녕하세요"))
```

- 자주 사용하는 문자 표현식

| 정규 표현식 | 설명 |
| ------- | ------ |
| `\d` | 숫자와 매치. [0-9]와 동일한 표현식 |
| `\D` | 숫자가 아닌 것과 매치. [^0-9]와 동일한 표현식 |
| `\s` | 화이트 스페이스(스페이스, 탭, 줄바꿈 등) 문자와 매치. [ \t\n\r\f\v]와 동일한 표현식 |
| `\S` | 화이트 스페이스(스페이스, 탭, 줄바꿈 등) 문자와 매치. [^ \t\n\r\f\v]와 동일한 표현식 |
| `\w` | 문자 + 숫자와 매치. [a-zA-Z0-9]와 동일한 표현식 |
| `\W` | 문자 + 숫자가 아닌 것과 매치. [^a-zA-Z0-9]와 동일한 표현식 |


```python
p = re.compile("[0-9]+")
p = re.compile(r"\d+")
print(p.match("012 abc"))

# 원본 문자열에서 0~9가 아닌 문자 찾기
#p = re.compile("[^0-9]+")
p = re.compile(r"\D+")
print(p.match("0123 abc"))


# 공백, 탭, 엔터 문자가 아닌 문자 찾기
p = re.compile(r"\S+")
print(p.search("0123 abc"))
```

```python
p = re.compile("[a-z]", re.I)

print(p.match("Python"))
print(p.match("Python"))
print(p.match("Python"))
```

```python
data = """python one 
life is too short 
python two 
you need python 
python three"""

# python으로 시작하고 '스페이스 하나', 영어, 숫자, _인 문자 찾기
p = re.compile(r"python\s\w+")
print(p.findall(data))
```

```python
# | : or

p = re.compile(r"Crow|Servo")
print(p.match("CrowHello"))
```

```python
# ^ : 시작하는, $ : 끝나는

p = re.compile("^Life")
m = p.search("Life is too Short")
print(m)

m = p.search("My Life")
print(m)

p = re.compile("short$")
m = p.search("Life is too short")
print(m)
```

```python
# () : 그루핑

p = re.compile("(ABC)+")
m = p.search("ABCABCABC K")
print(m)
print(m.group(1))
```

```python
# 이름 010-1234-5678

p = re.compile(r"(\w+)\s+(\d{3}-\d{4}-\d{4})")
m = p.search("park 010-1234-5678")
print(m)
print(m.group())
print(m.group(1))
print(m.group(2))
```

```python
# 찾아서 바꾸기
# John, Smith
# Smith, John

p = re.compile(r"(\w+)\s(\w+)")
m = p.search("John Smith")
print(m)

# re.sub(패턴, 바꿀 문자열, 원본 문자열)
re.sub(p,r"\2 \1","John Smith")
```

```python
# re.compile()
# re.sub()

# re.fullmatch() : 문자열 전체가 패턴과 일치

p = re.compile(r"\d+")

print(re.fullmatch(p, "123abc"))
print(re.fullmatch(p, "123456"))
print(re.fullmatch(p, "abc123"))

m = re.fullmatch(p,"abc123")
print(m.group())
```

```python
# 입력을 받은 값이 숫자인지 확인

# 숫자입니다 or 숫자가 아닙니다
# fullmatch()

p = re.compile(r"\d+")
num = input("숫자")

if re.fullmatch(p, num) :
    print("숫자입니다") 
else :
    print("숫자가 아닙니다")
```

```python
# 아이디, 비밀번호, 전화번호 입력
# 패턴에 맞을 때까지 입력받기

# 아이디 검사 조건
# 영문 대문자로 시작, 영문, 숫자, 특수문자(!@#$) 포함, 6~15자

# 비밀번호 검사 조건
# 영문 대소문자, 숫자, 특수문자(!@#$) 포함 8자 이상

# 전화번호 검사 조건
# 010-1234-5678 허용, 01012345678 허용
# fullmatch()
import re
deadcount = 5

while True:
    if deadcount <= 0 :
        break
    deadcount -= 1
    print(deadcount)
    p = re.compile(r"^[A-Z][\w!@#$]{5,14}$")
    id = input("아이디 입력")
    if re.fullmatch(p,id) :
        break

while True:
    if deadcount <= 0 :
        break
    deadcount -= 1
    print(deadcount)
    p = re.compile(r"[\w!@#$]{8,}$")
    password = input("비밀번호 입력")
    if re.fullmatch(p,password) :
        break

while True:
    if deadcount <= 0 :
        break
    deadcount -= 1
    print(deadcount)
    p = re.compile(r"^(010)\d{8}|(-\d{4}-\d{4})$")
    phone = input("전화번호 입력")
    if re.fullmatch(p,phone) :
        break

if deadcount > 0 :
    print(f"아이디 : {id}")
    print(f"비밀번호 : {password}")
    print(f"전화번호 : {phone}")
else :
    print("입력 오류")
```

```python
# 로그 파일에서 필요한 정보만 추출하기
# log.txt
# ERROR만 출력
# INFO 개수, ERROR 개수 출력
# 날짜별 ERROR 발생횟수 출력

logs = []
dates = {}
p_err = re.compile("ERROR")
p_info = re.compile("INFO")
p_date = re.compile(r"^(\d{4}-\d{2}-\d{2})")
err_cnt, info_cnt = 0, 0
try:
    with open("./data/log.txt","r") as f :
        logs = f.readlines()
except FileNotFoundError as e:
    print("파일 오류", e)

for log in logs :
    # 로그 개수 체크
    if p_err.search(log) :
        err_cnt += 1
        print("ERROR LOG :",log)
    elif p_info.search(log) :
        info_cnt += 1
        # print("INFO LOG :",log)

    # 날짜별 ERROR 로그 체크
    m = p_date.match(log)
    if m and p_err.search(log) :
        date = m.group(1)
        if date in dates :
            dates[date] += 1
        else :
            dates[date] = 1


print(f"INFO 개수 : {info_cnt}")
print(f"ERROR 개수 : {err_cnt}")
print(f"날짜별 ERROR 통계\n{dates}")
```
